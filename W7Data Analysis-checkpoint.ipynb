```python
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris)

plt.legend(bbox_to_anchor=(1, 1), loc=2)
```

```python
sns.scatterplot(x='petal_length', y='petal_width', hue='species', data=iris)
```

```python
iris_melted = pd.melt(iris,
                    id_vars='species',
                    value_vars=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'],
                    var_name='feature',
                    value_name='value')

sns.barplot(data=iris_melted, x='feature', y='value', hue='species', errorbar=None)

plt.title('Average Sepal and Petal Measurements by Species')
plt.xlabel('Feature')
plt.ylabel('Measurement (cm)')
plt.legend(title='Species')
```

```python
boxplot = sns.catplot(data=iris_melted, x='species', y='value', col='feature',
                kind='box', col_wrap=2, height=4, aspect=1.2)

boxplot.fig.suptitle('Boxplots of Iris Features by Species', y=1.03)
```

```python
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species')

plt.title('Sepal Length vs Sepal Width by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend(title='Species')
```

```python
sns.histplot(data=iris, x='petal_length', hue='species', kde=True, bins=20)

# Add title and labels
plt.title('Distribution of Petal Length Among Species')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
```
